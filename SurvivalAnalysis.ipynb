```r
library(survival)
library(ggplot2)
```

```r
# Read data
df <- read.csv('Datasets/heart_attack.csv')
```

```r
# Convert and confirm Date data type
df$admitdate <- as.Date(df$admitdate,'%m/%d/%Y')
df$disdate <- as.Date(df$disdate,'%m/%d/%Y')
df$fdate <- as.Date(df$fdate,'%m/%d/%Y')
```

```r
# A special visulization package from github
install.packages("remotes")
remotes::install_github("michaelway/ggkm")
```

```r
install.packages('ggkm')
library(ggkm)
```

```r
attach(df)
```

```r
# Create and plot a baseline survival model
km1 <- survfit(Surv(lenfol,fstat)~1)
summary(km1)
plot(km1, xlab = "Time", ylab = "Survival Probability")
```

```r
# Compare survival rate by gender
km2 <- survfit(Surv(lenfol,fstat)~gender)
summary(km2)
plot(km2, xlab = "Time", ylab = "Survival Probability")
ggplot(df, aes(time = lenfol, status = fstat, color = factor(gender)))+geom_km()
```

```r
# Compare survival rate by cvd
km3 <- survfit(Surv(lenfol,fstat)~cvd)
summary(km3)
plot(km3, xlab = "Time", ylab = "Survival Probability")
ggplot(df, aes(time = lenfol, status = fstat, color = factor(cvd))) + geom_km()
```

```r
# Using Proportional Hazards to test whether an attribute is significant in survival
coxph <- coxph(Surv(lenfol,fstat)~cvd, method="breslow")
summary(coxph)
```

```r
# To test all independent variables
X <- cbind(age,gender,hr,sysbp,diasbp,bmi,cvd,afb,sho,chf,av3,miord,mitype,los)
coxph <- coxph(Surv(lenfol,fstat)~X, method="breslow")
summary(coxph)
```
